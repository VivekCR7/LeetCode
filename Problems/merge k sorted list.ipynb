Given an array of linked-lists lists, each linked list is sorted in ascending order.

Merge all the linked-lists into one sort linked-list and return it.

Input: lists = [[1,4,5],[1,3,4],[2,6]]
Output: [1,1,2,3,4,4,5,6]
Explanation: The linked-lists are:
[
  1->4->5,
  1->3->4,
  2->6
]
merging them into one sorted list:
1->1->2->3->4->4->5->6

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

def Takeinput():
    inputlist = [int(ele) for ele in input().split()]
    head = tail = None
    
    for curr_data in inputlist:
        if curr_data == -1:
            break
        
        newnode = Node(curr_data)
        
        if head is None:
            head = newnode
            tail = newnode
            
        else:
            tail.next = newnode
            tail = newnode
            
    return head

# def mergeKlist(klists):
#     if not klists:
#         return

#     if len(klists) == 1:
#         return klists[0]

#     result = []

#     for node in klists:
#         while node:
#             result.append(node.data)
#             node = node.next

#     result.sort()


#     head = curr = Node()

#     for data in result:
#         curr.next = Node(data)
#         curr = head.next

#     return head.next

    
import heapq
def mergeKlist(klists):
    heap = []
    
    for node in klists:
        while node:
            heap.append(node.data)
            node = node.next
            
    heapq.heapify(heap)
    
    curr = head = Node(0)
    
    while heap:
        curr.next = Node(heapq.heappop(heap))
        curr = curr.next
        
    return head.next
    
    
    
def printLL(head):
    while head is not None:
        if head.next:
            print(str(head.data)+"->",end="")
        else:
            print(str(head.data)+".",end="")
        head = head.next

k = int(input())            
Klists = []
for l in range(k):
    l = Takeinput()
    Klists.append(l)
head = mergeKlist(Klists)
printLL(head)
```
