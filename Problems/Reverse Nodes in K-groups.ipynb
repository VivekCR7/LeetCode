```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
def Takeinput():
    inputlist = [int(ele) for ele in input().split()]
    head = tail = None
    for curr_data in inputlist:
        if curr_data == -1:
            break
            
        newnode = Node(curr_data)
        
        if head is None:
            head = tail = newnode
        else:
            tail.next = newnode
            tail = newnode
            
    return head

def printLL(head):
    while head is not None:
        if head.next:
            print(str(head.data)+"->",end="")
        else:
            print(str(head.data)+".",end="")
        head = head.next
        
    return


def reverseKGroup(head,k):
    current = head  
    next  = None
    prev = None
    count = 0 

    while(current is not None and count < k): 
        next = current.next
        current.next = prev 
        prev = current 
        current = next 
        count += 1
        
        
    if next is not None: 
        head.next = reverseKGroup(next, k) 


    return prev

def reverseKGroup1(head,k):
    if k == 1:
        return head
    def reverse(head, end):
        next = head.next
        rest = end.next
        head.next = rest
        while head != end:
            temp = next.next
            next.next = head
            head = next
            next = temp
        return head
    prev = None
    start = head
    while start is not None:
        end = start
        i = 1
        while end is not None and i < k:
            end = end.next
            i += 1
        if end is None:
            break
        rev = reverse(start, end)
        if prev is None:
            head = rev
        else:
            prev.next = rev
        prev = start
        start = start.next
    return head 
    
    
    

head = Takeinput()
k = int(input())
head = reverseKGroup1(head,k)
printLL(head)
    
    
    
```
